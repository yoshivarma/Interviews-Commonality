```python
import spacy
nlp = spacy.load('en_core_web_sm')
all_stopwords = nlp.Defaults.stop_words

doc1 = nlp('I saw the girl in the park yesterday. She was holding a green umbrella and was wearing an orange top.\
I saw her pass by in a red car. I was at home sitting by the window.')
doc2 = nlp('I think I saw the girl in an orange shirt yesterday. She had an umbrella on her and seemed to sit willingly in a car\
 that came around 5 o\'clock. I have not seen her since.')
docs = [doc1, doc2]

# Update car
noun_adj_pairs = []
for doc in docs:
  doc_noun_adj_pairs = []
  for chunk in doc.noun_chunks:
    phrase = ''
    noun = ''
    for tok in chunk:
          if tok.pos_ == "NOUN":
            noun = tok.text
            phrase = phrase + str(tok) + ' '
          if tok.pos_ == "ADJ":
            phrase = phrase + str(tok) + ' '
    if noun:
          if(len(phrase) > 0):
            phrase = phrase[0:-1]
          doc_noun_adj_pairs.append(phrase)
  noun_adj_pairs.append(doc_noun_adj_pairs)
  

print(noun_adj_pairs)
```

```python
import pandas as pd
import spacy
nlp = spacy.load('en_core_web_sm')

df = pd.read_csv("boulders_reviews.csv")

noun_adj_pairs = []

for ind in df.index:
  doc = df['review_text'][ind]
  if(type(doc) == type('')):
    doc = nlp(doc)
    doc_noun_adj_pairs = []
    for chunk in doc.noun_chunks:
      phrase = ''
      noun = ''
      for tok in chunk:
            if tok.pos_ == "NOUN":
              noun = tok.text
              phrase = phrase + str(tok) + ' '
            if tok.pos_ == "ADJ":
              phrase = phrase + str(tok) + ' '
      if noun:
            if(len(phrase) > 0):
              phrase = phrase[0:-1]
            doc_noun_adj_pairs.append(phrase)
    noun_adj_pairs.append(doc_noun_adj_pairs)
  
print(noun_adj_pairs)
```
